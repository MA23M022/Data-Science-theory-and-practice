```python
# 1.Write a program (WAP) to implement the following operations on a collection of library books:


# Q.1.a:- Construct a catalog of books, with each book having an author's name,
#     book title, ISBN number, publication year, and number of pages.

author_name = ['AA', 'BB', 'CC']
book_title = ['dd', 'ee', 'ff']
ISBN_num = [1024, 1034, 1024]
pub_year = [2010, 2000, 2020]
num_pages = [230, 340, 450]

print('The books in library:')
books = []
for ele in zip(author_name, book_title, ISBN_num, pub_year, num_pages):
    books.append(list(ele))
print(books)


# Q.1.b:- Add a new book to the catalog, ensuring that the books are kept in ascending order based on the publication year.

new_book = ['XX', 'yy', 2045, 2021, 500]
books.append(new_book)
print('Adding new book to the library:')
print(books)

for ele in books:
    dummy = ele[0]
    ele[0] = ele[3]
    ele[3] = dummy

books.sort()

for ele in books:
    dummy = ele[0]
    ele[0] = ele[3]
    ele[3] = dummy
    
print('After sorting by the year of publication:')
print(books)

# Q.1.c:- Locate a book by its ISBN number and delete the book's entry from the catalog.

Isbn = int(input('Enter the isbn number:'))
index = -1
for i in range(len(books)):
    if(books[i][2] == Isbn):
        index = i
        break        
if(index >= 0):
    print('The specified book is present at',index,'-th index and we goning to remove.')
    books.remove(books[index])
    print('After deleting the book, we have:')
    print(books)
else:
    print('There is no such book')
    
    
# Q.1.d:- Insert a new book entry at the end of the catalog using the provided book information.

new_book1 = ['FF', 'ss', 1024, 2012, 245]
pubs = new_book1[3]
for i in range(len(books)):
    if(pubs<books[i][3]):
        books.insert(i, new_book1)
        break
print('After inserting a new book:')
print(books)


# Q.1.e:- Identify and remove any duplicate entries in the catalog,
#           preserving only one copy of each book based on its ISBN number.


index1 = 0
while(index1 < len(books)):
    index2 = index1 + 1
    while(index2 < len(books)):
        if(books[index1][2] == books[index2][2]):
            books.remove(books[index2])
        else:
            index2 +=1
    index1 += 1
print('The updated library is:')
print(books)


# Q.1.f:- Reorganize the entire catalog so that the books are sorted in descending order by the number of pages.

for ele in books:
    dummy = ele[0]
    ele[0] = ele[4]
    ele[4] = dummy

books.sort()
books.reverse()

for ele in books:
    dummy = ele[0]
    ele[0] = ele[4]
    ele[4] = dummy
    
print('After sorting by number of pages:')
print(books)
```

```python
# 3. Using the zip function, WAP

mat1 = [[2, 3], [4, 5]]
mat2 = [[1, 2], [3, 4]]

# Q.3.a:- To add the elements of 2 matrices (Define matrices as per your wish).

mat3 = []
for ele1, ele2 in zip(mat1, mat2):
    val = []
    for ele3, ele4 in zip(ele1, ele2):
        val.append(ele3+ele4)
    mat3.append(val)
print('The desired addition output is:', mat3)

#Q.3.b:- To perform element wise multiplication on 2 matrices.

mat4 = []
for ele1, ele2 in zip(mat1, mat2):
    val = []
    for ele3, ele4 in zip(ele1, ele2):
        val.append(ele3*ele4)
    mat4.append(val)
print('The desired multiplication output is:', mat4)
```

```python
# 4. List of List : Given a square matrix represented as a list of lists, 


r, c = input('Enter the number of rows and columns of a matrix:').split()
r = int(r)
c = int(c)
print('The entered number of rows and columns is:', r, c)

mat = []
print('Enter the matrix:')
for i in range(r):
    print('Enter the', i+1,'-th row')
    val = []
    for j in range(c):
        ele = float(input())
        val.append(ele)
    mat.append(val)
print('The entered matrix is:', mat)

# Q.4.a:- WAP to print the row sum, column sum and trace of the matrix 

lst1 = []
for i in range(r):
    value = sum(mat[i])
    lst1.append(value)
print('The row sum list is:', lst1)

lst2 = []
for i in range(c):
    value = 0
    for j in range(r):
        value = value + mat[j][i]
    lst2.append(value)
print('The column sum list is:', lst2)

trace = 0
if(r == c):
    for i in range(r):
        trace += mat[i][i]
    print('The trace is:', trace)
else:
    print('The number of rows and columns should be same')

    
# Q.4.b:-  WAP to print the transpose of the matrix.

tran_mat = []
for i in range(c):
    val = []
    for j in range(r):
        val.append(mat[j][i])
    tran_mat.append(val)
print('The transpose matrix is:', tran_mat)

# Q.4.c:- WAP to check whether the given matrix is symmetric or not.

if(mat == tran_mat):
    print('The matrix is symmetric.')
else:
    print('The matrix is not symmetric.')
    
# Q.4.d:- WAP to check whether the Identity matrix (I) is positive definite or not by using
#         Quadratic form method (x^T*I*x > 0), where x is any non zero vector.


ide_mat = []
length = int(input('Enter the order of the identity matrix'))
for i in range(length):
    val = []
    for j in range(length):
        if(i == j):
            val.append(1)
        else:
            val.append(0)
    ide_mat.append(val)
print('The identity matrix is:', ide_mat)   

x = []
print('Enter the non-zero vector:')
for i in range(length):
    ele = float(input())
    x.append(ele)
print('The entered vector is:', x)

quad_val = 0
for i in range(length):
    quad_val += ide_mat[i][i]*(x[i]**2)
print('The calculated value is:', quad_val)
if(quad_val > 0):
    print('The identity matrix is positive definite.')
else:
    print('The identity matrix is not positive definite.')
```
