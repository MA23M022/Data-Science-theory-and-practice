# Q.1 Design a class named Polygon that initializes with the length of a side. 
Then, derive a class named Square from the Polygon class. Utilize the side length defined in the Polygon class for the Square class. Within the Square class, implement a method called findArea() that calculates and returns the square's area based on its side length. Use __ init __() for necessary initialization

```python
class Polygon:
    def __init__(self, s_l = 0):     
        self.side_length = s_l
        
    def show_side(self):
        print(f'The length of a side is {self.side_length}')
        
class Square(Polygon):     # inherits the Base class polygon.
    def __init__(self, side_length):
        super().__init__(side_length)
        # super() is a built-in function used to call methods defined in the parent class.
        
    def findArea(self):
        val = self.side_length**2
        return val
    
a1 = Square(5)  #creating a object of class Square.
a1.show_side()
area = a1.findArea()
print(f'The area of the square is {area}')
```

# 2.  (a)  Create a class Father with attributes

  - father_name (string), father_age (int), father_talents (set of strings)

    Create a class Mother with attributes:

  - mother_name (string), mother_age (int), mother_talents (set of strings)

    Create a class Child that inherits both father and mother with attributes: 

  - child_name (string), child_age (int), child_gender(string)

(b) Create a function getChildDetails() in Child to input the details of the child, it’s father and mother and printChildDetails() to print the details using                   printChildDetails()

(c) Create an object of class Child and read the details by invoking            getChildDetails() and display the details entered.

 (d) Create a function displayTalents() in class Child that displays the talents of the child inherited from father and mother. A talent is inherited to a child if both the parents possess it.

```python
class Father:   
    def __init__(self):
        self.father_name = None     # attributes are initialized by none.
        self.father_age = None
        self.father_talent = None
        
    def show_Father_Details(self):    # shows the details of the father.
        print(f'Father name = {self.father_name}')
        print(f'Father age = {self.father_age}')
        print(f'Father talent = {self.father_talent}')
        
class Mother:
    def __init__(self):
        self.mother_name = None
        self.mother_age = None
        self.mother_talent = None 
    
    def show_Mother_Details(self):        # shows the detail of the mother.
        print(f'Mother name = {self.mother_name}')
        print(f'Mother age = {self.mother_age}')
        print(f'Mother talent = {self.mother_talent}')
        
class Child(Father, Mother):       # the devied class child inherits Father and Mother base class.
    def __init__(self):
        self.child_name = None
        self.child_age = None
        self.child_gender = None
        self.child_talents = None
        
    def get_Child_Details(self, child_name, child_age, child_gender, f_n, f_a, f_t, m_n, m_a, m_t):
        self.father_name = f_n           # To assigned the attributes of the derived class.
        self.father_age = f_a
        self.father_talent = f_t
        self.mother_name = m_n
        self.mother_age = m_a
        self.mother_talent = m_t 
        self.child_name = child_name
        self.child_age = child_age
        self.child_gender = child_gender
        
    def print_Child_Details(self):     # print the child details.
        print(f'Child name = {self.child_name}')
        print(f'Child age = {self.child_age}')
        print(f'Child gender = {self.child_gender}')
        self.show_Father_Details()
        self.show_Mother_Details()
        
    def Display_talents(self):      # display the talent of the child which is possesed by father and mother.
        talent1 = list(self.father_talent)
        talent2 = list(self.mother_talent)
        
        child_talent = []
        for ele1 in talent1:
            for ele2 in talent2:
                if(ele1 == ele2):
                    child_talent.append(ele1)
                    break
                    
        child_talent = set(child_talent)
        print('The inherited talent is:', child_talent)

Father_talents = {'Reading book', 'Tv mechanics'}
Mother_talents = {'Reading book', 'Activist'}

c1 = Child()      # creating an object of child class.
c1.get_Child_Details('Sourav', 25, 'M', 'Kanai', 56, Father_talents, 'Shila', 47, Mother_talents)
c1.print_Child_Details()

c1.Display_talents()   # printing talent of the child.
```

#  3.  Text File Input Output

  Create a .txt (text) file and use the pledge of India as the content of the text file.

  Write a python program that reads this text file, processes it by counting the number of occurrences of each word in the file, and then writes the result back to a new text file.

```python
msg1 = """All Indians are my brothers and sisters.
I love my country and I am proud of its rich and varied heritage.
I shall strive to be worthy of it.
I shall respect my parents, teachers and all elders and treat everyone with courtesy."""

with open('India.txt', 'w') as text:    # create the text file to write the messege.
    text.write(msg1)
    
with open('India.txt', 'r') as text1:    # read the text file to make the list of words.
    data = text1.read()
    data = data.lower().split()

print('These are the list of words present in the text:')
print(data)


lst1 = []
lst2 = []

i = 0
while(i<len(data)):        # counting the number of occurrences of each word in the file.
    j = i + 1
    count = 1
    while(j<len(data)):
        if(data[i] == data[j]):
            count += 1
            data.pop(j)
        else:
            j += 1
    lst1.append(data[i])
    lst2.append(count)
    i += 1
    
with open('occurence.txt', 'w') as text2:  # creating the occurence text file to write the occurences.
    for ele1, ele2 in zip(lst1, lst2):
        val = ele1 + ' : ' + str(ele2)
        text2.write(val)
        text2.write('\n')
        
with open('occurence.txt', 'r') as text3:
    data1 = text3.read()
    print(data1)
```

Create a child class ‘ 'Restaurant Bill’ which has a property called ‘Price list’ of the items and has a method to calculate the total bill amount by using the price list and order details. Also have a   method to print the complete bill for the customer including taxes and service charges.

```python
class Bill:     # creating the bill class to find the details of the order.
    def __init__(self, customer_name, table_number, order):  
        self.customer_name = customer_name
        self.table_number = table_number
        self.order = order
        
    def order_detail(self):
        lst = self.order.replace(',', ' ')  # spiltting with respect to the comma(,). We get list of strings.
        lst = lst.split()
        
        lst2 = []
        for ele in lst:
            lst3 = []
            val = ele[-1: : -1]
            i = 0
            num = ''
            while(val[i] != 'x'):   # splitting with repect to ('x'). We get list of the items and their number.
                num += val[i]
                i += 1
            item = val[i+1 : ]
            lst3 = [item[-1: : -1], int(num[-1 : : -1])]
            lst2.append(lst3)
        return lst2
    
class Resturant_Bill(Bill):   # derived class which inherits base class Bill.
    def __init__(self, customer_name, table_number, order, price_list):
        super().__init__(customer_name, table_number, order)
        self.price_list = price_list
        
    def total_bill(self):     # computing total bill by the item.
        lst = self.price_list.replace(',', ' ')
        lst = lst.split()
        
        lst2 = []
        for ele in lst:
            lst3 = []
            val = ele[-1: : -1]
            i = 0
            num = ''
            while(val[i] != '-'):  # spillting with respect to ('-'). we get the list of items and their price.
                num += val[i]
                i += 1
            item = val[i+1 : ]
            lst3 = [item[-1: : -1], int(num[-1 : : -1])]
            lst2.append(lst3)
            
        lst4 = super().order_detail()  
        bill = 0
        for ele1, ele2 in zip(lst4, lst2):
            bill += (ele1[1]*ele2[1])
        return bill
        
    def complete_bill(self):  # compute complete bill including taxes and service charges.
        bill = self.total_bill()
        c_bill = bill + (bill*0.15) + (bill*0.1)
        return c_bill
        

order = "Item1x1,Item2x3,Item3x1"
price_list = "Item1-100,Item2-70,Item3-250"
b1 = Resturant_Bill('Sourav', 5, order, price_list)   # create an object of Resturant_Bill class.

bill = b1.total_bill()
print('The total bill by the items is:', bill)

complete_bill = b1.complete_bill()
print('The complete bill by resturant is:', complete_bill)
```
