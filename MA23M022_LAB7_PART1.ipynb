# 1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

```python
import numpy as np
import matplotlib.pyplot as plt

A, B, C, D, E = input('Enter the cofficient of the fourth order polyniminal').split()
A = float(A)
B = float(B)
C = float(C)
D = float(D)
E = float(E)
print(f'The entered co-efficients are {A}, {B}, {C}, {D}, {E}')

x = np.linspace(-100, 100, 20)  # creates an array of evenly spaced values over a specified interval
y = A*pow(x, 4) + B*pow(x, 3) + C*pow(x, 2) + D*pow(x, 1) + E

print('The x values are:')
print(x)

print('The corresponding polynominal values are:')
print(y)

plt.figure(figsize=(6,6))
plt.plot(x, y, 'o-', color='red', lw = 1.5, ms = 2, label = 'x-val. ploy-val.')
plt.xlabel('x-values')
plt.ylabel('polynominal values')
plt.title('The graph of x-value vs corresponding value')
plt.legend(loc='upper right', fontsize = 10)
plt.show()
```

# 2. Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.

```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = {

'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

file = pd.DataFrame.from_dict(sales_data)
print(file)

total_sale = file.groupby('Product')['Sales'].sum()  # using groupby() function.
print(total_sale)

Product_Name = list(total_sale.index)
print('The name of the products are:', Product_Name)
Product_amount = list(total_sale.values)
print('The amount of products are:', Product_amount)
 
# creating the bar plot
fig = plt.figure(figsize = (10, 5))
plt.bar(Product_Name, Product_amount, color ='maroon',  width = 0.5)
plt.xlabel("Product names")
plt.ylabel("Product amounts")
plt.title("Total sale for each product")
plt.show()
```

# 4. For the function J(w) = w^2 + (54/w), implement the bracketing method (choose your own a, b, n).

```python
def J(w):
    val = w**2 + (54/w)
    return val

def bracketing_method(a, b, n):
    dw = (b - a)/n
    w1 = a
    w2 = w1 + dw
    w3 = w2 + dw
    while(w3 <= b):
        if((J(w1) >= J(w2)) and (J(w2) <= J(w3))):
            print(f'Minimum lies between ({w1}, {w3})')
            return
        else:
            w1 = w2
            w2 = w3
            w3 = w3 + dw
    print(f'Minimum does not lie in ({a}, {b}). Min could be one of the bdry points.')
    return
           
a, b = input('Enter the interval:').split()
a = int(a)
b = int(b)

n = int(input('Enter the number of subintervals: '))
bracketing_method(a, b, n)
```
