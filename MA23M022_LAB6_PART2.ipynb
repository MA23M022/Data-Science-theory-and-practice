# 6.    Create a base class Vehicle with the following attributes:

 make (string)

 model (string)

 year (int)

 Create a method initialize_vehicle to set the above attributes. Also, create a method display_vehicle to print these    attributes.

 Create a class Car inherited from Vehicle with the following additional attribute:

 fuel_type (string)

 Create a method get_car_details to initialize the above attribute along with Vehicle attributes.

 Also, create a method display_vehicle to print these attributes along with Vehicle attributes.

 Create a class Bike inherited from Vehicle with the following additional attribute:

 gear_count (int)

 Create a method get_bike_details to initialize the above attribute along with Vehicle attributes.

 Also, create a method display_vehicle to print these attributes along with Vehicle attributes.

 Create two different objects for Car and Bike and demonstrate each of the methods.

 Example -1:

 my_car = Car()

 my_car.get_car_details("Toyota", "Camry", 2020, "Petrol")

 my_car.display_vehicle()

 Output:

 Make: Toyota, Model: Camry, Year: 2020

 Fuel Type: Petrol

 Example -2 :

 my_bike = Bike()

 my_bike.get_bike_details("Yamaha", "YZF R1", 2021, 6)

 my_bike.display_vehicle()

 Output:

 Make: Yamaha, Model: YZF R1, Year: 2021

 Gear Count: 6

```python
class Vehicle:           # create Vehicle base class.
    def __init__(self):
        self.make = None
        self.model = None
        self.year = None
        
    def initialize_vehicle(self, make, model, year):   # This member function assigns value to the attributes.
        self.make = make
        self.model = model
        self.year = year
        
    def display_vehicle(self):        # prints whatever assigned to the attributes.
        print('Make:', self.make)
        print('Model:', self.model)
        print('Year:', self.year)
        
class Car(Vehicle):       # Car derived class inherits the vehicle base class.
    def __init__(self):
        self.fuel_type = None
    
    def get_car_details(self, make, model, year, fuel_type):   # This function assigns value to the attributes .
        self.fuel_type = fuel_type
        Vehicle.initialize_vehicle(self, make, model, year)
        
    def display_vehicle(self):
        Vehicle.display_vehicle(self)
        print('Fuel type:', self.fuel_type)

class Bike(Vehicle):    # Bike derived class inherits the vehicle base class.
    def __init__(self):
        self.gear_count = None
    
    def get_bike_details(self, make, model, year, gear_count):
        self.gear_count = gear_count
        Vehicle.initialize_vehicle(self, make, model, year)
        
    def display_vehicle(self):         # This function displays the bike details.
        Vehicle.display_vehicle(self)
        print('Fuel type:', self.gear_count)
        
print('-----:CAR:-----')  
my_car = Car()         # create an object of the Car class.
my_car.get_car_details( "Toyota", "Camry", 2020, "Petrol")
my_car.display_vehicle()

print(" ")
print('----:BIKE:----')  
my_bike = Bike()       # create an object of the bike class.
my_bike.get_bike_details("Yamaha", "YZF R1", 2021, 6)
my_bike.display_vehicle()
```

# 8. Class Inheritance in Python: Finding GCD (greatest common divisor) and LCM (least common multiple) of Numbers and Handling Composite Numbers. 

a) Create a Numbers class with a, b, find_gcd(), and find_lcm() methods.

b) Create an EvenNumbers class that inherits from Numbers and overrides find_lcm() to handle even numbers.

c) Create an OddNumbers class that inherits from Numbers and overrides find_lcm() to handle odd numbers.

d) Create a CompositeNumbers class that inherits from EvenNumbers and OddNumbers and overrides find_gcd() to handle composite numbers.

e) Create a CompositeNumbers object with a = 12 and b = 9, and call its find_lcm() and find_gcd() methods.

```python
import math as m

class Numbers:
    def __init__(self, num1, num2):
        if((num1 >= 0) and (num2 >= 0)):   # checks the numbers are non-negetive or not.
            self.a = num1
            self.b = num2
            self.decision = True
        else:
            self.decision = False
        
    def find_gcd(self):                 # find the gcd of two numbers.
        if(self.decision):
            val1 = m.gcd(self.a, self.b)
            print(f'The gcd of {self.a} and {self.b} is {val1}')
        else:
            print('The given numbers should be non-negetive')
            
    def find_lcm(self):               # find the lcm of two numbers.
        if(self.decision):
            val2 = (self.a * self.b) / m.gcd(self.a, self.b)
            print(f'The lcm of {self.a} and {self.b} is {val2}')
        else:
            print('The given numbers should be non-negetive')
            
class EvenNumbers(Numbers):       # EvenNumber derived class inherits the Number base class.
    def __init__(self, num1, num2):
        Numbers.__init__(self, num1, num2)
        
    def find_lcm(self):         # overrides find_lcm() to handle even numbers.
        if(self.decision):
            if(((self.a % 2) == 0) and ((self.b % 2) == 0)):
                val2 = (self.a * self.b) / m.gcd(self.a, self.b)
                print(f'The lcm of {self.a} and {self.b} is {val2}')
            else:
                print('The numbers are not even numbers')
                super().find_lcm()     # call the base class member function find_lcm()
        else:
            print('The given numbers should be non-negetive')
            
class OddNumbers(Numbers):     # OddNumber derived class inherits the Number base class.
    def __init__(self, num1, num2):
        Numbers.__init__(self, num1, num2)
        
    def find_lcm(self):       # find_lcm() to handle odd numbers.
        if(self.decision):
            if(((self.a % 2) == 1) and ((self.b % 2) == 1)):
                val2 = (self.a * self.b) / m.gcd(self.a, self.b)
                print(f'The lcm of {self.a} and {self.b} is {val2}')
            else:
                print('The numbers are not odd numbers')
                super().find_lcm()   # call the base class member function find_lcm()
        else:
            print('The given numbers should be non-negetive')
            
class CompositeNumbers(EvenNumbers, OddNumbers):   # composite derived class which inherits EvenNumbers, OddNumbers class.
    def __init__(self, num1, num2):
        EvenNumbers.__init__(self, num1, num2)
        OddNumbers.__init__(self, num1, num2)
        
    def find_gcd(self):       # overrides find_gcd() to handle composite numbers.
        if(self.decision):
            if((m.gcd(self.a, self.b)) != 1):
                val = m.gcd(self.a, self.b)
                print(f'The gcd of {self.a} and {self.b} is {val}')
            else:
                print('The numbers are not composite numbers and we call base class member function to find gcd.')
                super().find_gcd()     # call the base class member function find_lcm(). 
        else:
            print('The given numbers should be non-negetive')
        
            
n1 = CompositeNumbers(12, 9)   # create an object of the CompositeNumbers class.
n1.find_gcd()
n1.find_lcm()
```

# 10. WAP to create a pandas dataframe with a list of words and sort them in ascending order. The sorted words should be copied to a new file.

```python
import pandas as pd

header = ['words']
data = ['Alice', 'Charlie', 'Bob']    # By dataframe member function of the pandas moudle, we assign the header columns to data.
data = pd.DataFrame(data, columns = header)   
data.to_csv('list_of_words.csv', index=False)  # By to_csv member function , we name the csv file as list_of_words.

data = pd.read_csv('list_of_words.csv')  # By the read_csv member function, we read the csv file.
dummy = list(data.words)
print(dummy)

dummy = sorted(dummy)    # sort the list.
print(dummy)

dummy = pd.DataFrame(dummy, columns = header)   # By the above procedure , we make the another csv file named
                                                  #as new_list_of_words
dummy.to_csv('new_list_of_words', index = False)
```
