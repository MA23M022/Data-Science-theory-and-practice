```python
# Q.1. Explore the use and syntax of common built-in functions: 

#(Q.1.a) range() :-

##  The Python range() function returns a sequence of numbers, in a given range.
##  stntax:  range(start, stop, step), where start and step values are optional.

lst = []
for i in range(6):
    lst.append(i)
print('Using range() function:', lst)

#(Q.1.b) iter() :- 

## iter() method returns the iterator object, it is used to convert an iterable to the iterator.

list1 = [1, 2, 3, 4, 5]
print(type(list1))
list1 = iter(list1)
print(type(list1))
print('The value from the list is:', next(list1))

#(Q.1.c) eval() :- 

## Python eval() function parse the expression argument and
## evaluate it as a Python expression and runs Python expression (code) within the program.

expression = 'x*(x+1)*(x+2)'
print(expression)

x = 3

result = eval(expression)
print('The required result is:', result)


#(Q.1.d) enumerate() :-

## Syntax: enumerate(iterable, start=0), by default it starts from index 0.
## Returns an iterator with index and element pairs from the original iterable.

lst2 = [3, 5, 7]
print('Using enumerate() function:')
for count, ele in enumerate(lst2):
    print(count, ele)

#(Q.1.e) zip():-

## Python zip() method takes iterable containers and returns a single iterator object,
##  having mapped values from all the containers. 

lst3 = [1, 3, 5]
lst4 = [12, 45, 67]
lst5 = []
for ele in zip(lst3, lst4):
    lst5.append(ele)
    
print('Using zip() function:', lst5)

#(Q.1.f) lambda :-

## Python Lambda Functions are anonymous functions means that the function is without a name.
## As we already know the def keyword is used to define a normal function in Python.
## Similarly, the lambda keyword is used to define an anonymous function in Python. 

Max = lambda a, b : a if(a > b) else b
print('Using lambda function:', Max(5, 2))

#(Q.1.g) input() :-

## This function is used for taking input from user. And it takes as string.

str1 = input('Enter the string: ')
print('The entered value is:', str1)

#(Q.1.h) map() :- 

## map() function returns a map object(which is an iterator)
##  of the results after applying the given function to each item of a given iterable (list, tuple etc.) 

# Double all numbers using map and lambda

numbers = (1, 2, 3, 4)
result = map(lambda x: x + x, numbers)
print('Using map() function:', list(result))

# filter() :- 

## The filter() method filters the given sequence with the help of a function that
##      tests each element in the sequence to be true or not. 

# a list contains both even and odd numbers. 
seq = [0, 1, 2, 3, 5, 8, 13]

# result contains odd numbers of the list
result = filter(lambda x: x % 2 != 0, seq)
print('Using filter function containing odd numbers:', list(result))

# next() :-

## Python’s next() function returns the next item of an iterator.

l = [1, 2, 3]
l_iter = iter(l) 
print('Using next() function:', next(l_iter))

# reduce() :-

## The reduce(fun,seq) function is used to apply a particular function passed in its argument
## to all of the list elements mentioned in the sequence passed along.This function is defined in “functools” module.


# importing functools for reduce() 
import functools 

# initializing list 
lis = [1, 3, 5, 6, 2] 

# using reduce to compute sum of list 
print("The sum of the list elements is : ", end="") 
print(functools.reduce(lambda a, b: a+b, lis)) 

# using reduce to compute maximum element from list 
print("The maximum element of the list is : ", end="") 
print(functools.reduce(lambda a, b: a if a > b else b, lis)) 
```

```python
# Q.2. Write a Python function that sorts a dictionary based on the length of values.

#Sample:

#Input: {'lemon':'yellow','apple':'red'} output: {'apple':'red','lemon':'yellow'}

def sort_length(dict1):
    lst = []
    for key, value in dict1.items():
        lst.append([key, value])
    for ele in lst:
        dummy = ele[0]
        ele[0] = len(ele[1])
        ele[1] = dummy
    lst.sort()
    for ele in lst:
        dummy = ele[0]
        ele[0] = ele[1]
        ele[1] = dummy
    dict2 = {ele[0]:dict1[ele[0]] for ele in lst}
    return dict2

dictionary = {'lemon':'yellow','apple':'red', 'orange':'orange'}
sorted_dict = sort_length(dictionary)
print('The sorted dictionary based on the length of values is', sorted_dict)
```

```python
# Q.3. Develop a Python program that executes the following tasks with a user-provided string:

# Q.3.a:- Prompt the user to input a string.

str1 = input('Enter a string:')

# Q.3.b:- Create a dictionary from the string where each key is a unique alphabet character and the corresponding
#  value is the frequency of that character's occurrence in the string.

lst1 = []
lst2 = []
for ele in str1:
    if ele in lst1:
        pass
    else:
        lst1.append(ele)
        lst2.append([ele, str1.count(ele)])
        
dict1 = dict(lst2)
print('The required dictionary is:', dict1)

#Q.3.c:- Generate a sorted list of tuples from the dictionary based on character frequency (values).

lst = []
for key, value in dict1.items():
        lst.append([key, value])
        
for ele in lst:
    dummy = ele[0]
    ele[0] = ele[1]
    ele[1] = dummy
    
lst.sort()
for ele in lst:
    dummy = ele[0]
    ele[0] = ele[1]
    ele[1] = dummy

for i in range(len(lst)):
    lst[i] = tuple(lst[i])
    
print('The sorted list of tuples from the dictionary based on character frequency (values) is:', lst)

# Q.3.d:- Generate a sorted list of tuples from the dictionary based on the alphabet characters (keys).

lst1 = []
for key in dict1.keys():
        lst1.append(key)
        
lst1.sort()
lst2 = [(ele, dict1[ele]) for ele in lst1]
print('The required sorted list of tuples from the dictionary based on the alphabet characters (keys):', lst2)

# Q.3.e. Identify the three most frequently occurring characters. In the event of a frequency tie,
#   prioritize characters in lexicographical order.

sorted_list = list(sorted(dict1.items(), key = lambda x:(-x[1], x[0])))
print('sorted list:', sorted_list[:3])
```

```python
# Q.4. Write a function called lookup_student that takes a dictionary representing student records,
#      where names are keys and roll numbers are values. The function should search for a specified student name
#       and return the corresponding roll number if found; otherwise, it should return "Not Found" 

# Example:

# records = { "Alice" : "AB111", "Bob" : "EE200", "David" : "XY333"}

# print(lookup_student(records, "Bob")) : Should print "EE200"

# print(lookup_student(records, "John")) : Should print "Not Found"

def lookup_student(records, name):
    for key in records.keys():
        if key == name:
            return records[key]
    return None

records = { "Alice" : "AB111", "Bob" : "EE200", "David" : "XY333"}
value = lookup_student(records, 'David')
if(value):
    print('The corrsponding roll number is:', value)
else:
    print('Not Found')
```
