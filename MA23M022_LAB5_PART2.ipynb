# 5. WAP to implement a class called "Bank_Account" representing a person's bank account.
The class should have the following attributes: account_holder_name (str), account_number(int), address (str) and balance (float).
The class should have methods to implement the following:
    deposit - Deposits a given amount into the account
    withdraw - Withdraws a given amount from the account
    check_balance - To get the current balance
    update_details - To update the name and address from the user and displays a message indicating successful update
    display_details - To display the details of the account.

```python
class Bank_Account:    # creating bank account class
    def __init__(self, ac_name = None, ac_number = None, address = None, balance = None): # constructor defined
        self.account_holder_name = ac_name
        self.account_number = ac_number
        self.address = address
        self.balance = balance
        
    def deposit(self, credit):      # defining deposit() member function
        if(self.account_number != None):
            self.balance += credit
            print(f'The balance is credited by {credit} and the current balance is {self.balance}')
        
    def withdraw(self, debit):     # defining withdraw() member function
        if(self.account_number != None):
            self.balance -= debit
            print(f'The balance is debited by {debit} and the current balance is {self.balance}')
        
    def check_balance(self):       # defining check_balance() member function
        if(self.account_number != None):
            print(f'The current balance is {self.balance}')
        
    def update_details(self, new_name, new_adress):  # defining update_details() member function
        if(self.account_number != None):
            self.account_holder_name = new_name
            self.address = new_adress
            print('Addresses are successfully updated')
        
    def display_details(self):           # defining display-details() member function 
        if(self.account_number != None):
            print(f'The account holder name is {self.account_holder_name}')
            print(f'The account number is {self.account_number}')
            print(f'The address of the account is {self.address}')
            print(f'The balance of the account is {self.balance}')
        else:
            print('Cannot find account holder. Give account details when create an object of the class.')
        
s1 = Bank_Account('Sourav', 2325, 'Contai', 12000)   # creating an object of Bank_Account class
s1.display_details()

s1.deposit(5000)
s1.withdraw(2000)
s1.check_balance()
s1.update_details('Sourav Majhi', 'Kolkata')

s1.display_details()
```

# 8.  Create a class named Geometry that provides static methods for various geometric calculations, such as area and perimeter, for different shapes (circle, rectangle, square). Include:
Static methods like circle_area(radius), rectangle_area(length, width), and square_area(side).
Static methods for perimeter calculations for the above shapes.
Ensure that methods check for valid inputs (e.g., positive numbers).

```python
# There can be some functionality that relates to the class,
#   but does not require any instance(s) to do some work, static methods can be used in such cases.
#     A static method is a method which is bound to the class and not the object of the class.
#       A static method does not receive an implicit first argument. 

from math import pi

class Geometry:
    @staticmethod
    
    def circle_area(radius):
        if(radius > 0):
            val = pi*(radius**2)
            return val
        else:
            return 'Radius of the circle should be positive value'
        
    def rectangle_area(length, width):
        if((length > 0)and(width > 0)):
            val = length*width
            return val
        else:
            return 'Length and Width of the rectangle should be positive value'
        
    def square_area(side):
        if(side > 0):
            val = side**2
            return val
        else:
            return 'Side of the square should be positive'
        
    def circle_perimeter(radius):
        if(radius > 0):
            val = 2*pi*radius
            return val
        else:
            return 'Radius of the circle should be positive value'
        
    def rectangle_perimeter(length, width):
        if((length > 0)and(width > 0)):
            val = 2*(length + width)
            return val
        else:
            return 'Length and Width of the rectangle should be positive value'
        
    def square_perimeter(side):
        if(side > 0):
            val = 4*side
            return val
        else:
            return 'Side of the square should be positive'

        
area1 = Geometry.circle_area(3)
print('The area of the circle is:',area1)

area2 = Geometry.rectangle_area(4, 6)
print('The area of the rectangle is:',area2)

area3 = Geometry.square_area(5)
print('The area of the square is:',area3)

perimeter1 = Geometry.circle_perimeter(3)
print('The perimeter of the circle is:',perimeter1)

perimeter2 = Geometry.rectangle_perimeter(4, 6)
print('The perimeter of the rectangle is:',perimeter2)

perimeter3 = Geometry.square_perimeter(5)
print('The perimeter of the square is:',perimeter3)
```
