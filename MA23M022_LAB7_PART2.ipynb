# 5. WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.arange(0, 10*np.pi, 0.08)
X = np.cos(t)
Y = np.sin(t)
Z = t

# Plot using scatter method
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')
scatter_plot = ax.scatter3D(X, Y, Z, color = 'green', label='Helical Wave Scatter')
ax.set_xlabel('X-axis', fontweight ='bold') 
ax.set_ylabel('Y-axis', fontweight ='bold') 
ax.set_zlabel('Z-axis', fontweight ='bold')

plt.legend(loc='upper right', fontsize = 15) # Add legend

plt.show() # Show plots
```

# 8.Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy

```python
import numpy as np

matrix = np.random.randint(0, 10, size = (3, 2))
print('The matrix is:')
print(matrix)

transpose_matrix = matrix.T
print('The transpose of the matrix is:')
print(transpose_matrix)

flat_matrix1 = matrix.flatten('C')
print('Flatten in row-major (C-style) order:')
print(flat_matrix1)

flat_matrix2 = matrix.flatten('F')
print('Flatten in column-major (Fortran-style) order:')
print(flat_matrix2)

flat_matrix3 = matrix.flatten('A')
print('Flatten in column-major order if matrix is Fortran contiguous in memory, row-major order otherwise')
print(flat_matrix3)

flat_matrix4 = matrix.flatten('K')
print('Flatten matrix in the order the elements occur in memory')
print(flat_matrix4)
```

# 9. WAP to capitalize a column of names in a Pandas Dataframe.

Eg : Input : {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

Output : {'Name': ['John', 'Boday', 'Ana', 'Peter', 'Nicky'], 'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}

```python
import pandas as pd

Input = {'Name': ['john', 'bODAY', 'aNa', 'Peter', 'nicky'],
         'Education': ['masters', 'graduate', 'graduate', 'Masters', 'Graduate'], 'Age': [27, 23, 21, 23, 24]}
    
data = pd.DataFrame.from_dict(Input)
print('Before capitalization of the column of names:')
print(data)

data['Name'] = data['Name'].str.capitalize()
print('\nAfter capitalization of the column of names:')
print(data)
```

# 10. Use the central difference method to find the first and second order derivatives of the function. 
Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of Optimization for Engineering Design: Algorithms and Examples by KALYANMOY DEB, 2nd edition

f(x) = 3x**2 + 2x

```python
def f(x):
    val = 3*(x**2) + (2*x)
    return val

def first_order_derivative(a, i, dx):
    left = a + (i-1)*dx
    right = a + (i + 1)*dx
    val = (f(right) - f(left))/(2*dx)
    return val
    
def second_order_derivative(a, i, dx):
    left = a + (i-1)*dx
    mid = a + i*dx
    right = a + (i + 1)*dx
    val = (f(right) - 2*f(mid) + f(left))/pow(dx, 2)
    return val

a = 4
b = 2
n = 10
dx = (b-a)/n

first = first_order_derivative(a, 5, dx)
print('The first order derivative is:', first)

second = second_order_derivative(a, 5, dx)
print('The second order derivative is:', second)
```
