```python
# Q.1. Write a program(WAP) using loops and recursion: 

#  a) Factorial of n where n is a non negative integer. 

def facto(n):
    if(n==0):
        return 1
    fact = n*facto(n-1)
    return fact

n = int(input('Enter the a non-negetive integer:'))
fact = 1
if(n==0):
    pass
else:
    for i in range(1, n+1):
        fact = fact*i
print('Using loops the factorial of n is:', fact)

fact1 = facto(n)
print('Using recursion the factorial of n is',fact1)

#  b) For calculating the Nth Fibonacci number.

def fibonaci(n):
    if((n == 1) or (n == 2)):
        return 1
    fibo = fibonaci(n-1) + fibonaci(n-2)
    return fibo

a = 1
b = 1
c = a+b
i = 3
while(i<n):
    a = b
    b = c
    c = a+b
    i += 1
    
print('Using loops the Nth Fibonacci number is:', c)

fibo1 = fibonaci(n)
print('Using recursion the Nth Fibonacci number is:', fibo1)

#  c) To calculate a^b where a>0, b>=0.

def power(a, b):
    if(b == 0):
        return 1
    value = a*power(a, b-1)
    return value

a = int(input('Enter a positive base:'))
b = int(input('Enter a non-negetive exponent:'))

val = 1
if(b == 0):
    pass
else:
    i = 1
    while(i <= b):
        val = val*a
        i += 1
print('Using loops the value of a^b is:', val)

value1 = power(a, b)
print('Using recursion the value of a^b is:', value1)
```

```python
# Q.2. Query for 2 integers N and M from the user where 0<=N<=100 and 0<=M<=9.
#      These will be the inputs to your function. Using recursion,  compute the number of times the integer M
#      occurs in all non-negative integers less than or equal to N.
#      example: For N=13 and M=1, count=6 (numbers 1,10,11,12,13).

def count_fun(N, M):
    if(N< 10): 
        return 1
    num = N
    quotient = num//10
    rem = num%10
    count = 0
    
    if((num == 100) and (M == 1)):
        count = count + 1
    if((num == 100) and (M == 0)):
        count = count + 1
        
    if(quotient == M):
        count = count + 1
    if(rem == M):
        count = count + 1
    return count + count_fun(N-1, M)

N = int(input('Enter a integer where 0<=N<=100 :'))
M = int(input('Enter a integer where 0<=N<=9 :'))

count1 = count_fun(N, M)
print('The number is:', count1)
```

```python
# Q.3. Programs using lambda function.

# a) Given a list of names, use map to create a list where each name is prefixed with "Hello, ".

#    Example Input: ['Alice', 'Bob', 'Charlie']
#    Example Output: ['Hello, Alice', 'Hello, Bob', 'Hello, Charlie']

str1 = 'Hello'
lst1 = ['Alice', 'Bob', 'Charlie']

result = map(lambda x : str1+', '+x, lst1)
print('The desired output using map() function is:', list(result))

# b) Use filter and a lambda function to extract all even numbers from a given list.

#    Example Input: [1, 2, 3, 4, 5, 6]
#    Example Output: [2, 4, 6]

lst1 = [1, 2, 3, 4, 5, 6, 7, 8]
lst2 = list(filter(lambda x: x%2 == 0, lst1))
print('Using filter() function:', lst2)

# c) Use reduce and lambda to concatenate all strings in a given list.

#    Example Input: ['Python', 'is', 'awesome']
#    Example Output: 'Pythonisawesome'

import functools
lst1 = ['Python', 'is', 'awesome']
print('Using reduce() function:',functools.reduce(lambda x, y: x+y, lst1))
```

```python
# Q.4.  Define a class Complex that defines a complex number with attributes real and imaginary (as we did in the class).
#       Define operators for addition, subtraction, multiplication and division (Do with both operator overloading as well
#       as without overloading). While printing the output, print in the form of complex number form like ( a + ib)  - 10 marks
#       ( 1 mark each for each of the operations with and without operator overloading)

class Complex:
    def __init__(self, r = 0, im = 0):
        self.real = r
        self.imaginary = im
        
    def addition(self, other):
        temp = Complex()
        temp.real = self.real + other.real
        temp.imaginary = self.imaginary + other.imaginary
        return temp
    
    def __add__(self, other):
        temp = Complex()
        temp.real = self.real + other.real
        temp.imaginary = self.imaginary + other.imaginary
        return temp
    
    def substraction(self, other):
        temp = Complex()
        temp.real = self.real - other.real
        temp.imaginary = self.imaginary - other.imaginary
        return temp
    
    def __sub__(self, other):
        temp = Complex()
        temp.real = self.real - other.real
        temp.imaginary = self.imaginary - other.imaginary
        return temp
    
    def multiplication(self, other):
        temp = Complex()
        temp.real = (self.real*other.real) - (self.imaginary*other.imaginary) 
        temp.imaginary = (self.real*other.imaginary) + (self.imaginary*other.real)
        return temp
    
    def __mul__(self, other):
        temp = Complex()
        temp.real = (self.real*other.real) - (self.imaginary*other.imaginary) 
        temp.imaginary = (self.real*other.imaginary) + (self.imaginary*other.real)
        return temp
    
    def divison(self, other):
        if(((other.real**2) + (other.imaginary**2)) == 0):
            print('Dividing by Zero. Please give vaid argument')
            return None
        
        temp = Complex()
        denominator = (other.real**2) + (other.imaginary**2)
        temp.real = round(((self.real*other.real) + (self.imaginary*other.imaginary))/denominator, 2)
        temp.imaginary = round(((self.imaginary*other.real) - (self.real*other.imaginary))/denominator, 2)
        return temp
    
    def __truediv__(self, other):
        if(((other.real**2) + (other.imaginary**2)) == 0):
            return None
        
        temp = Complex()
        denominator = (other.real**2) + (other.imaginary**2)
        temp.real = round(((self.real*other.real) + (self.imaginary*other.imaginary))/denominator, 2)
        temp.imaginary = round(((self.imaginary*other.real) - (self.real*other.imaginary))/denominator, 2)
        return temp
    
    def show(self):
        val = (lambda x : '+' if(self.imaginary >= 0) else '-')(self.imaginary)
        print(f'{self.real} {val} {abs(self.imaginary)}i')

print('The first complex number is:', end = ' ')
c1 = Complex(2, 3)
c1.show()

print('The second complex is:', end = ' ')
c2 = Complex(4, 5)
c2.show()

print('The addition without operator overloading is:', end = ' ')
c3 = Complex()
c3 = c1.addition(c2)
c3.show()

print('The addition with operator overloading is:', end = ' ')
c4 = Complex()
c4 = c1 + c2
c4.show()

print('The substraction without operator overloading is:', end = ' ')
c5 = Complex()
c5 = c1.substraction(c2)
c5.show()

print('The substrction with operator overloading is:', end = ' ')
c6 = Complex()
c6 = c1 - c2
c6.show()

print('The multiplication without operator overloading is:', end = ' ')
c7 = Complex()
c7 = c1.multiplication(c2)
c7.show()

print('The multiplication with operator overloading is:', end = ' ')
c8 = Complex()
c8 = c1*c2
c8.show()

print('The divison without operator overloading is:', end = ' ')
c9 = Complex()
c9 = c1.divison(c2)
if(c9):
    c9.show()
else:
    print('Dividing by Zero. Please give vaid argument')

print('The divison with operator overloading is:', end = ' ')
c11 = Complex()
c10 = Complex()
c10 = c1/c2
if(c10):
    c10.show()
else:
    print('Dividing by Zero. Please give vaid argument')
```
